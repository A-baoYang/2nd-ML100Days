了解如何使用 Sklearn 中的 hyper-parameter search 找出最佳的超參數

### 作業
請使用不同的資料集，並使用 hyper-parameter search 的方式，看能不能找出最佳的超參數組合

```python
from sklearn import metrics, datasets
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.ensemble import GradientBoostingRegressor, GradientBoostingClassifier
```

```python
digits = datasets.load_digits()

x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=5)

GDBT = GradientBoostingClassifier(random_state=7)

GDBT.fit(x_train, y_train)

y_pred = GDBT.predict(x_test)

print('MSE: ', metrics.mean_squared_error(y_test, y_pred))
```

```python
n_estimator = [100, 200, 300]
max_depth = [1, 3, 5]
param_grid = dict(n_estimators=n_estimator, max_depth=max_depth)

grid_search = GridSearchCV(GDBT, param_grid, scoring='neg_mean_squared_error', n_jobs=1, verbose=1)

grid_result = grid_search.fit(x_train, y_train)
```

```python
print('Best Accuracy: %f using %s' % (grid_result.best_score_, grid_result.best_params_))
```

```python
GDBT_best_param = GradientBoostingClassifier(max_depth=grid_result.best_params_['max_depth'],
                                             n_estimators=grid_result.best_params_['n_estimators'])

GDBT_best_param.fit(x_train, y_train)

y_pred = GDBT_best_param.predict(x_test)
```

```python
print('MSE: ', metrics.mean_squared_error(y_test, y_pred))
```
