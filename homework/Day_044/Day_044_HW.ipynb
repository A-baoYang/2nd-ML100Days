```python
from sklearn import metrics, datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
wine = datasets.load_wine()

x_train, x_test, y_train, y_test =train_test_split(wine.data, wine.target, test_size=0.2, random_state=5)

RFC = RandomForestClassifier(n_estimators=20, max_depth=4)

RFC.fit(x_train, y_train)

y_pred = RFC.predict(x_test)
```

```python
print('Accuracy Score: ', metrics.accuracy_score(y_test, y_pred))
print('Feature Importance: ', RFC.feature_importances_)
```

```python
from sklearn.ensemble import RandomForestRegressor

boston = datasets.load_boston()

x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.2, random_state=5)

RFR = RandomForestRegressor(n_estimators=30, max_depth=5)

RFR.fit(x_train, y_train)

y_pred = RFR.predict(x_test)
```

```python
print('Accuracy Score: ',RFR.score(x_test, y_test)) #https://www.kaggle.com/nsrose7224/random-forest-regressor-accuracy-0-91
print('Feature Importance: ', RFR.feature_importances_)
```
