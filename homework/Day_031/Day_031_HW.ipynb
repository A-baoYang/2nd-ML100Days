## 廣告投放優化的前置決策
### 原因
1. 目標明確：例如，想要點擊成本在幾塊以內
2. 重複性高：例如，再行銷受眾事件（加購物車、瀏覽商品幾日...）會有一固定架構先走，再根據數據優化
3. 有結果回饋可再餵給機器學習：例如，蒐集所有圖像對各種受眾版位下的成本和轉換表現，讓機器學習對新素材、受眾、版位組合給出成效預測。
