```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
wine = datasets.load_wine()
boston = datasets.load_boston()
```

```python
wine.target
```

```python
boston.target
```

```python
# 篩其中1個feature(column)來看
wine_X = wine.data[:, np.newaxis, 6]
print('Data Shape: ', wine_X.shape)

# 切分訓練集/測試集
wine_X_train, wine_X_test, wine_y_train, wine_y_test = train_test_split(wine_X, wine.target, test_size=0.2, random_state=5)

# 因目標是分類類型，建一個邏輯回歸模型
logi = linear_model.LogisticRegression()
logi.fit(wine_X_train, wine_y_train)
wine_y_pred = logi.predict(wine_X_test)
```

```python
acc = accuracy_score(wine_y_test, wine_y_pred)
print(acc)
```

```python
boston_X = boston.data[:,np.newaxis, 5]
print('Data Shape: ', boston_X.shape)

boston_X_train, boston_X_test, boston_y_train, boston_y_test = train_test_split(boston_X, boston.target, test_size=0.1, random_state=5)

regr = linear_model.LinearRegression()
regr.fit(boston_X_train, boston_y_train)
boston_y_pred = regr.predict(boston_X_test)
```

```python
print('Coeffients: ', regr.coef_)
print('Mean Squared Error: ', mean_squared_error(boston_y_test, boston_y_pred))
```

```python
# 畫回歸模型與實際資料分析
plt.scatter(boston_X_test, boston_y_test, color='blue')
plt.plot(boston_X_test, boston_y_pred, color='green')
plt.show()
```
