```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, 150, 151)
w = 7.6
b = 0.9

y = (x + np.random.randn(151) * 5) * w + b

plt.plot(x, y, 'b.', label = 'data points')
plt.legend(loc = 2)
plt.show()
```

```python
y_fakepredict = x * w + b

plt.plot(x, y, 'b.', label = 'data points')
plt.plot(x, y_fakepredict, 'r-', label = 'fake prediction')
plt.legend(loc = 2)
plt.show()
```

```python
def mean_square_error(y, yp):
    """
    計算 MSE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mse: MSE
    """
    power_yyp = (y - yp)**(2)
    mse = MSE = sum(power_yyp) / len(y)
    return mse
```

```python
MSE = mean_square_error(y, y_fakepredict)
print("The Mean Square Error is %.3f" % (MSE))
```
