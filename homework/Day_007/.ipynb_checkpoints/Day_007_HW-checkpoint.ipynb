# 處理 outliers
* 新增欄位註記
* outliers 或 NA 填補
    1. 平均數 (mean)
    2. 中位數 (median, or Q50)
    3. 最大/最小值 (max/min, Q100, Q0)
    4. 分位數 (quantile)

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '../../data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 1. 列出 AMT_ANNUITY 的 q0 - q100

```python
# 1: ignore NA, 計算 AMT_ANNUITY 的 q0 - q100
q100_num = list(range(101))
quantile_100s = [np.percentile(app_train[~app_train['AMT_ANNUITY'].isnull()]['AMT_ANNUITY'], q = i) for i in q100_num]
```

```python
pd.DataFrame({'q': q100_num,
              'value': quantile_100s})
```

## 2.1 將 AMT_ANNUITY 中的 NAs 暫時以中位數填補

```python
# 2.1 將 NAs 以 q50 填補
print("Before replace NAs, numbers of row that AMT_ANNUITY is NAs: %i" % sum(app_train['AMT_ANNUITY'].isnull()))
```

```python
q_50 = np.percentile(app_train[~app_train['AMT_ANNUITY'].isnull()]['AMT_ANNUITY'], q = 50)
app_train.loc[app_train['AMT_ANNUITY'].isnull(),'AMT_ANNUITY'] = q_50
print("After replace NAs with q50=%d, numbers of row that AMT_ANNUITY is NAs: %i" % (q_50, sum(app_train['AMT_ANNUITY'].isnull())))
```

## 2.2 將 AMT_ANNUITY 的數值標準化至 -1 ~ 1 間

### Hints: Normalize function (to -1 ~ 1)
$ y = 2*(\frac{x - min(x)}{max(x) - min(x)} - 0.5) $

```python
# 2.2 Normalize values to -1 to 1
print("== Original data range(已將NA填補成中位數的) ==")
print(app_train['AMT_ANNUITY'].describe())

def normalize_value(x):
    value = x.values
    normal_x = ((value - min(value)) / (max(value) - min(value)) - 0.5) * 2
    return normal_x

app_train['AMT_ANNUITY_NORMALIZED'] = normalize_value(app_train['AMT_ANNUITY'])

print("== Normalized data range ==")
app_train['AMT_ANNUITY_NORMALIZED'].describe()
```

```python
app_train['AMT_ANNUITY_NORMALIZED'].hist(bins = 50)
plt.title("AMT_ANNUITY - Normalized to -1 ~ 1")
plt.show()
```

## 3. 將 AMT_GOOD_PRICE 的 NAs 以眾數填補

```python
print("Before replace NAs, numbers of row that AMT_GOODS_PRICE is NAs: %i" % sum(app_train['AMT_GOODS_PRICE'].isnull()))
```

```python
app_train['AMT_GOODS_PRICE'].value_counts()
```

```python
# scipy計算眾數
from scipy.stats import mode
value_most = mode(app_train[~app_train['AMT_GOODS_PRICE'].isnull()]['AMT_GOODS_PRICE'])
print(value_most)

# 一般計算眾數方法
mode_goods_price = list(app_train['AMT_GOODS_PRICE'].value_counts().index)
app_train.loc[app_train['AMT_GOODS_PRICE'].isnull(), 'AMT_GOODS_PRICE'] = mode_goods_price[0]

print("After replace NAs with mode=%d, numbers of row that AMT_GOODS_PRICE is NAs: %i" % (mode_goods_price[0], sum(app_train['AMT_GOODS_PRICE'].isnull())))
```

```python
# 其他計算眾數方法
from collections import defaultdict
mode_dict = defaultdict(lambda: 0)
for value in app_train[~app_train['AMT_GOODS_PRICE'].isnull()]['AMT_GOODS_PRICE']:
    mode_dict[value] += 1

value_most = sorted(mode_dict.items(), key=lambda kv: kv[1], reverse=True)
app_train.loc[app_train['AMT_GOODS_PRICE'].isnull(), 'AMT_GOODS_PRICE'] = value_most[0][0]
print("After replace NAs with mode=%d, numbers of row that AMT_GOODS_PRICE is NAs: %i" % (value_most[0][0], sum(app_train['AMT_GOODS_PRICE'].isnull())))
```
