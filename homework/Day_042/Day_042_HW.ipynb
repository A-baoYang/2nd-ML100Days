```python
from sklearn import datasets, metrics

# 如果是分類問題，請使用 DecisionTreeClassifier，若為回歸問題，請使用 DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
wine = datasets.load_wine()

x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.1, random_state=5)

DTC = DecisionTreeClassifier()

DTC.fit(x_train, y_train)

y_pred = DTC.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print('accuracy score: ', acc)
```

```python
print(wine.feature_names)
```

```python
print('Feature Importance: ', DTC.feature_importances_)
```

```python
DTC_entropy = DecisionTreeClassifier(criterion='entropy')
```

```python
DTC_entropy.fit(x_train, y_train)
y_pred = DTC_entropy.predict(x_test)
print('Accuracy Score: ', metrics.accuracy_score(y_test, y_pred))
print('Feature Importance: ', DTC_entropy.feature_importances_)
```
