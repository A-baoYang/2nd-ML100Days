## 開始檢視不同欄位是否有異常值

```python
# DAYS_BIRTH: 客戶申請貸款時的年齡
(application_train_df['DAYS_BIRTH'] / (-365)).describe()
```

```python
# DAYS_EMPLOYED: 申請貸款前，申請人已在現職工作的時間
(application_train_df['DAYS_EMPLOYED'] / 365).describe()
plt.hist(application_train_df['DAYS_EMPLOYED'])
plt.show()
application_train_df['DAYS_EMPLOYED'].value_counts()
```

### 從上面的圖與數值可以看出, 365243 顯然是個奇怪的數值

```python
anom = application_train_df[application_train_df['DAYS_EMPLOYED'] == 365243]
non_anom = application_train_df[application_train_df['DAYS_EMPLOYED'] != 365243]
print('The non-anomalies default on %0.2f%% of loans' % (100 * non_anom['TARGET'].mean()))
print('The anomalies default on %0.2f%% of loans' % (100 * anom['TARGET'].mean()))
print('There are %d anomalous days of employment' % len(anom))
```

```python
# 新增一個欄位: DAYS_EMPLOYED_ANOM 來標記 DAYS_EMPLOYED 是否異常
application_train_df['DAYS_EMPLOYED_ANOM'] = application_train_df["DAYS_EMPLOYED"] == 365243
print(application_train_df['DAYS_EMPLOYED_ANOM'].value_counts())

# 這邊我們用 nan 將異常值取代
application_train_df['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

application_train_df['DAYS_EMPLOYED'].plot.hist(title = 'Days Employment Histogram')
plt.xlabel('Days Employment')
```
